```python
class tartaruga():
    def recebe_lab(self, N):
        self.lab = N
    def iniciar(self): #optei por não colocar uma função __init__ pois demanda a existência de um labirinto
        self.ite = 0
        self.a0 = self.lab.shape[0]
        self.b0 = self.lab.shape[1]
        self.x , self.y = random.randint(0,self.a0-1) , random.randint(0,self.b0-1)
        while self.lab[self.x, self.y] == 1:
            self.x , self.y = random.randint(0,self.a0) , random.randint(0,self.b0)
        self.lab[self.x, self.y] = 2
        self.dentro=True
    def posicao(self):
        print('('+str(self.x)+','+str(self.y)+')')
        return self.x, self.y
    def saiu(self):
        if self.x == 0 or self.y==0 or self.x==self.lab.shape[0]-1 or self.x==self.lab.shape[1]-1:
            self.dentro=False
            print('Parabéns, você saiu!')
    def andar(self, ordem):
        if ordem=='N':
            if self.lab[self.x-1, self.y] == 0 and self.dentro:
                self.lab[self.x, self.y] = 0
                self.x-=1
                self.lab[self.x, self.y] = 2
                self.ite += 1
                self.saiu()
                return True
        elif ordem=='S':
            if self.lab[self.x+1, self.y] == 0 and self.dentro:
                self.lab[self.x, self.y] = 0
                self.x+=1
                self.lab[self.x, self.y] = 2
                self.ite += 1
                self.saiu()
                return True
        elif ordem=='L':
            if self.lab[self.x, self.y+1] == 0 and self.dentro:
                self.lab[self.x, self.y] = 0
                self.y+=1
                self.lab[self.x, self.y] = 2
                self.ite += 1
                self.saiu()
                return True
        elif ordem=='O':
            if self.lab[self.x, self.y-1] == 0 and self.dentro: 
                self.lab[self.x, self.y] = 0
                self.y-=1
                self.lab[self.x, self.y] = 2
                self.ite += 1
                self.saiu()
                return True
        else:
            return False
    def tenta_sair(self):
        estrategia = [['N', 'S', 'L', 'O'],['S', 'L', 'O', 'N'], ['L','O','N','S'], ['O','N','S','L']]
        lugares = [(self.x, self.y)]
        est = 0
        dire = 0
        while self.dentro==True and self.ite<10000:
            a = self.andar(estrategia[est][dire])
            if a:
                if self.posicao() in lugares:
                    est = (est+1)%4
                    print(lugares)
                    lugares=[]
                else:
                    lugares.append(self.posicao())
            else:
                dire = (dire+1)%4
        if self.dentro:
            print('Impossivel sair')
            print(self.ite)
        else:
            print('Saiu')
```

O problema de Monty Hall, também conhecido por paradoxo de Monty Hall é um problema matemático e paradoxo que surgiu a partir de um concurso televisivo dos Estados Unidos chamado "Let’s Make a Deal", exibido na década de 1970. O jogo consiste no seguinte: Monty Hall (o apresentador) apresentava 3 portas aos concorrentes, sabendo que atrás de uma delas está um carro (prêmio bom) e que as outras têm prêmios de pouco valor.

- Na 1ª etapa o concorrente escolhe uma porta (que ainda não é aberta);  
- Em seguida, Monty abre uma das outras duas portas que o concorrente não escolheu, sabendo de antemão que o carro não se encontra na porta que abrirá;  
- Agora com duas portas apenas para escolher — pois uma delas já se viu, na 2ª etapa, que não tinha o prêmio — e sabendo que o carro está atrás de uma delas, o concorrente tem que se decidir se permanece com a porta que escolheu no início do jogo e abre-a ou se muda para a outra porta que ainda está fechada para então a abrir.  

Demonstre graficamente que é mais provável ganhar o prêmio se trocar de porta do que se não o fizer. Quais as probabilidade de acertar, em ambos os casos?

```python
def jogosemtroca(): #código simplificado
    portasorteada=random.randint(1,3)
    portaescolhida=random.randint(1,3)
    if portasorteada==portaescolhida:
        return True
    else:
        return False
```
