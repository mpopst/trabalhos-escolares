# Data Cleaning

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random as r
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap
from matplotlib import cm
```

```python
A = pd.read_csv("survey_results_public.csv")
```

```python
A.head() #Just to see the df
```

```python
#We are changing the variables to English. X and Y are auxiliary variables.
X = A['Hobby'].replace(to_replace='Yes',value='Sim')
Y = X.replace(to_replace='No', value='Não')
A['Hobby'] = Y
X = A['OpenSource'].replace(to_replace='Yes',value='Sim')
Y = X.replace(to_replace='No', value='Não')
A['OpenSource'] = Y
A.head()
```

```python
#Changing by the average and filling the small countries.
A.ConvertedSalary = A.groupby(['Country'])['ConvertedSalary'].apply(lambda x: x.fillna(x.mean()))
A.ConvertedSalary = A['ConvertedSalary'].fillna(0)
```

```python
#Creating the collum MonthlySalary
A['MonthlySalary'] = (1/12)*A['ConvertedSalary']
```

```python
A['BrazilMonthlySalary'] = 3.81*A['MonthlySalary']
```

```python
#Foo and bar are auxiliary variables. The method describe works at every numeric 
#collum. We use this fact to create aux, the list with every numeric collum name.
foo = list(A.describe()) #every numeric collum
bar = list(A) #every collum
aux = [item for item in bar if item not in foo] #BAR - FOO: every non-numeric
A[aux] = A[aux].fillna('Unknown')
```

# Plots

```python
fig, ax = plt.subplots()

X = A[A['Country']=='Brazil']['BrazilMonthlySalary']
Y = A[A['Country']=='United States']['BrazilMonthlySalary']
Z = A[A['Country']=='India']['BrazilMonthlySalary']
br_heights, br_bins = np.histogram(X)
usa_heights, usa_bins = np.histogram(Y, bins=br_bins)
ind_heights, ind_bins = np.histogram(Z, bins=br_bins)

width = (br_bins[1] - br_bins[0])/3

ax.bar(br_bins[:-1], br_heights, width=width, facecolor='seagreen')
ax.bar(usa_bins[:-1]+2*width, usa_heights, width=width, facecolor = 'red')
ax.bar(ind_bins[:-1]+width, ind_heights, width=width, facecolor = 'orange')
ax.legend(['Brazil','India','USA'])
plt.xlabel('Salário, em milhares de reais')
plt.ylabel('Quantidade de programadores')
```

```python
X = A.groupby('Country')['BrazilMonthlySalary'].mean()
X.sort_values(inplace=True, ascending=True)
X.plot(kind='barh',figsize=(12,27),grid=True,title='Average Salary for every country, in Brazilian Real')
```

```python
b = A[(A['Country'] != 'Unknown') & (A['FormalEducation'] != 'Unknown')]
c = b[['BrazilMonthlySalary','Country','FormalEducation']].groupby(['FormalEducation','Country'],as_index=False).mean()
c = c[(c['FormalEducation'] == 'Bachelor’s degree (BA, BS, B.Eng., etc.)') | 
      (c['FormalEducation'] == 'Master’s degree (MA, MS, M.Eng., MBA, etc.)') |
      (c['FormalEducation'] ==  'Other doctoral degree (Ph.D, Ed.D., etc.)')]
x = c.groupby(['Country']).mean()
x.columns = ['Salario Alto Grau']
c = b.groupby(['Country'],as_index=False).mean()[['Country','BrazilMonthlySalary']]
resp = x.merge(c,how='left',on='Country')

resp1 = resp[(resp['Country'] == 'Brazil') | (resp['Country'] == 'India') |(resp['Country'] == 'United States') 
             |(resp['Country'] == 'Venezuela, Bolivarian Republic of...')]
resp1.columns = ['Country','Salario Alto Grau', 'Media']
resp1.groupby('Country').mean().plot(kind='bar')
```

```python
A['JobSatisfaction'] = A['JobSatisfaction'].replace({
                              'Extremely dissatisfied':-3,'Extremely satisfied':3, 
                             "Moderately dissatisfied":-2, "Moderately satisfied":2,
                              "Slightly dissatisfied":-1, "Slightly satisfied":1,
                             "Neither satisfied nor dissatisfied":0,
                             "Unknown":np.nan})
A['JobSearchStatus'] = A['JobSearchStatus'].replace({
                            "I am actively looking for a job":-1,
                            "I’m not actively looking, but I am open to new opportunities":0,
                            "I am not interested in new job opportunities":1,
                            "Unknown":np.nan})
b = A[['JobSearchStatus','JobSatisfaction']].dropna()
b['aux'] = 1
```

```python
colors = sns.color_palette("coolwarm_r", n_colors=7)
cmap1 = LinearSegmentedColormap.from_list("my_colormap", colors)
c = b.groupby(['JobSearchStatus','JobSatisfaction']).count().unstack('JobSatisfaction')
c['aux'].plot.bar(stacked=True, figsize=(20, 15), rot=1, colormap=cmap1, title='Satisfação por job status')
```

Após as breves alterações, é fácil perceber que existe esta correlação. A satisfação no trabalho aumenta muito conforme o status do trabalho. Veja que a quantidade de pessoas felizes no terceiro caso é muito maior e a quantidade de pessoas infelizes, proporcionalmente, no primeiro caso, é muito maior. Dito isto, é possivel inferir tal correlação.

```python
A['LastNewJob'] = A['LastNewJob'].replace({
                            "Between 1 and 2 years ago":1.5,
                            "Between 2 and 4 years ago":3,
                            "I've never had a job":0,
                            "Less than a year ago":0.5,
                            "More than 4 years ago": 6, "Unknown":np.nan})
b = A[['JobSatisfaction','LastNewJob']].dropna()
b['aux'] = 1
```

```python
colors = sns.color_palette("coolwarm_r", n_colors=7)
cmap1 = LinearSegmentedColormap.from_list("my_colormap", colors)
c = b.groupby(['LastNewJob','JobSatisfaction']).count().unstack('JobSatisfaction')
c['aux'].plot.bar(stacked=True, figsize=(20, 15), rot=1, colormap=cmap1, title = 'Satisfação por tempo de carreira')
```

É curioso porque a quantidade de pessoas insatisfeitas permanece quase que constante, no entanto, há uma tendência de queda das pessoas satisfeitas. Nos faz conjecturar que provavelmente existe um grupo considerável que só permanece no emprego se estiver satisfeito (em torno de $20\%$ dos recém contratados, na matemática freestyle). Provavelmente os tais millenials, que trocam de emprego igual trocam de roupa.

Também nos leva a crer que existe um pessoal que sempre vai estar insatisfeito com o trabalho. Eles ficam insatisfeitos mas não mudam de emprego nem nada.

Isso significa que as empresas devem oferecer um bom ambiente de trabalho logo no começo, para evitar demissões em massa.

```python
meanGERAL = A[A['Country'] != 'United States']['BrazilMonthlySalary'].mean()
meanUSA = A[A['Country'] == 'United States']['BrazilMonthlySalary'].mean()
RESP = (meanUSA/meanGERAL - 1)*100
print('A média de salário nos EUA é %.2f%% maior que no resto do mundo.' % RESP)
```

```python
b = pd.DataFrame(A.LanguageWorkedWith.str.split(';').tolist(), index=A.Respondent)
b = b.reset_index()[[0, 'Respondent']] 
b.columns = ['Respondent', 'Worked']
b = b.groupby('Respondent').count()

c = pd.DataFrame(A.LanguageDesireNextYear.str.split(';').tolist(), index=A.Respondent)
c = c.reset_index()[[0, 'Respondent']] 
c.columns = ['Respondent', 'Desired']
c = c.groupby('Respondent').count()

B = b.merge(c, left_on='Respondent',right_on='Respondent')
B['Percent'] = B['Desired'] / B['Worked']
X = B['Percent'].sort_values(inplace=False, ascending=True)
X.plot(kind='barh',figsize=(12,14),grid=True,title='Languages people want to learn next year compared with now.')
```

```python
b = pd.DataFrame(A.DatabaseWorkedWith.str.split(';').tolist(), index=A.Respondent)
b = b.reset_index()[[0, 'Respondent']] 
b.columns = ['Respondent', 'Worked']
b = b.groupby('Respondent').count()

c = pd.DataFrame(A.DatabaseDesireNextYear.str.split(';').tolist(), index=A.Respondent)
c = c.reset_index()[[0, 'Respondent']] 
c.columns = ['Respondent', 'Desired']
c = c.groupby('Respondent').count()

B = b.merge(c, left_on='Respondent',right_on='Respondent')
B['Percent'] = B['Desired'] / B['Worked']
X = B['Percent'].sort_values(inplace=False, ascending=True)
X.plot(kind='barh',figsize=(12,14),grid=True,title='Database people want to learn next year compared with now.')
```
