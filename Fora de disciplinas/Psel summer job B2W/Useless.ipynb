```python
import pandas as pd
```

```python
df = pd.read_csv("base_de_respostas_10k_amostra.csv")
```

```python
list(df)
```

```python
A = df[["Gender", "ConvertedSalary","YearsCoding"]]
```

```python
A = A.dropna()
```

```python
A["Gender"] = A["Gender"].replace("Male",0)
A["Gender"] = A["Gender"].replace("Female",1)
A = A[(A["Gender"] == 1) | (A["Gender"] == 0)]
```

```python
A.groupby("Gender").count()
```

```python
men = A[A["Gender"] == 0]
men = men.sample(n=373)
```

```python
women = A[A['Gender'] == 1]
```

```python
res = pd.concat([women,men])
```

```python
res = res.sample(frac=1)
```

```python
res
```

```python
res.groupby("YearsCoding").count()
```

```python
res["AnosProgramando"] = res["YearsCoding"].replace({"0-2 years":1,"12-14 years":13,"15-17 years":16,"18-20 years":19,"21-23 years":22,
                                      "24-26 years":25,"27-29 years":28,"3-5 years":4,"30 or more years":33,"6-8 years":7,
                                     "9-11 years":10})
```

```python
res = res[["Gender","ConvertedSalary","AnosProgramando"]]
```
